**9.1**

**a)** Consider vectors in $\mathcal{R}^2$:

$$\mathbf{A} = \begin{bmatrix}2 \\ 1\end{bmatrix}; \ \ 
\mathbf{B} = \begin{bmatrix}3 \\ 4\end{bmatrix}; \ \ 
\mathbf{C} = \begin{bmatrix}1 \\ 3\end{bmatrix}; \ \ 
\mathbf{D} = \begin{bmatrix}6 \\ 8\end{bmatrix}; \ \ 
\,.$$

Suppose a linear transformation maps $\mathbf{A}$ to $\mathbf{C}$, $\mathbf{B}$ to  $\mathbf{D}$, and conserves the origin; does this describe it fully? Write out the transformation.


Mapping: $\mathcal{M}(2,1) = (1,3)

$\mathcal{M}(3,4) = 6,8)$

$\mathcal{M}(0,0) = (0,0)$

x' = ax + by + c

y' = dx + ey + f

So what are our a, b ,c ,d?

The first set of coordinates for points A and C = M(A) gives us: 

x' = 1 = ax + by = a2 + b1

y' = 3 = cx + dy = c2 + d1

The second set:

x' = 6 = ax + by = a3 + b4

y' = 8 = cx + dy = c3 + d4

So this is a matrix equation 
AX = V

a*2 + b*1 + c*0 + d*0 = 1
c*2 + d*1 + a*0 + b*0 = 3
a*3 + b*4 + c*0 + d*0 = 6
c*3 + d*4 + a*0 + b*0 = 8




```python
from numpy import array
from numpy.linalg import solve

A = array([[2,1,0,0],[0,0,2,1],[3,4,0,0],[0,0,3,4]],float)
V = array([1,3,6,8])

X = solve(A,V)
print(X.reshape(4,1))

print("The thransformation is: x' =", X[0], "x +", X[1], "y\n")
print(" y' =", X[2],"x +", X[3], "y\n")
```

```python
from matplotlib.pyplot import plot, show, gca, rcParams, xlim, ylim
from numpy import cos, sin, pi

ax = gca()
ax.set_aspect('equal')
xlim(-15,15)
ylim(-15,15)
rcParams['figure.figsize'] = [20, 20]

D = array([[2,1],[0,0],[3,4],[9,2],[7,1],[5,3],[2,1]],float)
print(D)

c15 = cos(15/180*pi)
s15 = sin(15/180*pi)

R2xy = array([[c15,-s15], [s15, c15]], float)

plot(D[:,0],D[:,1])

Dp = (R2xy@D.T).T
plot(Dp[:,0],Dp[:,1])

Dpp = (R2xy@Dp.T).T
plot(Dpp[:,0],Dpp[:,1])

Dppp = (R2xy@Dpp.T).T
plot(Dppp[:,0],Dppp[:,1])

Dpppp = (R2xy@Dppp.T).T
plot(Dpppp[:,0],Dpppp[:,1])

Dppppp = (R2xy@Dpppp.T).T
plot(Dppppp[:,0],Dppppp[:,1])

temp = (R2xy@Dppppp.T).T
plot(temp[:,0],temp[:,1])

for i in range(20):
    tempy = (R2xy@temp.T).T
    plot(tempy[:,0],tempy[:,1])
    temp = tempy
show()
```
