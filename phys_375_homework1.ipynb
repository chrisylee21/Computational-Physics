```python
#7
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def equation(x):
    return (x * ((np.exp(x)-1)**-1) * np.exp(x)) - 5

x_guess = 4.75
x_solution = fsolve(equation, x_guess)

print(f"The solution for x is: {x_solution[0]:.10f}")

x_values = np.linspace(3, 7, 500) 
y_values = equation(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="Derivative of Planck's Law", color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--', label="y = 0")
plt.scatter(x_solution, equation(x_solution), color='red', label=f"Solution x ≈ {x_solution[0]:.5f}", zorder=5)
plt.title("Plot of the Derivative of Planck's Law and its Solution")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#7
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

h = 6.626e-34 
c = 2.998e8  
k = 1.38e-23  

T = 5772 # Using the temperature of the Sun       

a = 2*h*c**2
b = h*c/k

def planck_law(wavelength, T):
    """Planck's law for blackbody radiation."""
    return (a*(wavelength**-5))/(np.exp(b/(wavelength*T)) - 1)

def planck_derivative(wavelength, T):
    """Derivative of Planck's law with respect to wavelength."""
    term1 = (a*b*np.exp(b/(wavelength*T)))/(T*(wavelength**7)*(((np.exp(b/(wavelength*T)))-1)**2))
    term2 = (-5*a)/((wavelength**6)*(np.exp(b/(wavelength*T))-1))
    return term1 + term2

wavelength_guess = 500e-9  
wavelength_solution = fsolve(planck_derivative, wavelength_guess, args=(T))[0]

print(f"Peak wavelength (λ_max): {wavelength_solution * 1e9:.2f} nm")

wien_constant = wavelength_solution * T
print(f"Wien's displacement constant (b): {wien_constant:.5e} m·K")

wavelengths = np.linspace(100e-9, 3000e-9, 1000)  
intensities = [planck_law(w, T) for w in wavelengths]

plt.figure(figsize=(8, 6))
plt.plot(wavelengths * 1e9, intensities, label=f"T = {T} K")
plt.axvline(wavelength_solution * 1e9, color='red', linestyle='--', label=f"λ_max = {wavelength_solution * 1e9:.2f} nm")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Spectral Radiance (W·m⁻²·sr⁻¹·m⁻¹)")
plt.title("Blackbody Radiation Spectrum")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#8a
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve


def equation(x):
    return 3*((np.exp(x))-1) - x*np.exp(x)

x_guess = 2.75

x_solution = fsolve(equation, x_guess)

print(f"The solution for x is: {x_solution[0]:.10f}")

x_values = np.linspace(0.5, 3.5, 500) 
y_values = equation(x_values)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="Derivative of Planck's Law", color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--', label="y = 0")
plt.scatter(x_solution, equation(x_solution), color='red', label=f"x ≈ {x_solution[0]:.5f}", zorder=5)
plt.title("Plot of the Derivative of Planck's Law")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
